**Extracting the data from this website**
https://www.education.gouv.fr/annuaire?keywords=&city=&department=&establishment=All&status=All&geo_point=

```python
import requests
from scrapy import Selector
import pandas as pd
import traceback
import json
```

```python
def scrap_etablissements_contact():
    results = {}
    
    def post_processing_string(string,field='optional'):
        if field == "tags":
            res = [tag.replace("\n","").strip() for tag in string]
            return res

        res = string.replace("\n","").strip()
        
        if field == "email":
            return res.replace("mailto:", "").strip()
        
        def keep_only_numerics(string):
            return "".join(c for c in string if c.isdigit())
        
        if field == 'telephone': 
            return keep_only_numerics(res)

        return res
    
    page = 0
    index = 0
    
    while True: 
        try: # iterate on pages, until there aren't any pages left
            html = requests.get('https://www.education.gouv.fr/annuaire?keywords=&city=&department=&establishment=All&status=All&geo_point=&page={}'.format(page)).content
            sel = Selector(text = html)
            etablissements = sel.xpath('//div[contains(@class,"etablissement etablissement--extra_search_result etablissement--search__item")]')

            # Parsing the page
            for i,etablissement in enumerate(etablissements):
                try:
                    nom_etablissement = etablissement.xpath('//div[contains(@class,"establishment--search_item__content")]/h2/text()').extract()[i]
                except Exception as e: 
                    print("no name")
                    print(e)
                    print(page)
                    print(i)
                    print("")
                    pass
                
                try:
                    address = etablissement.xpath('//div[contains(@class,"establishment--search_item__content")]/p/text()').extract()[i*3]
                except Exception as e: 
                    print("no address")
                    print(e)
                    print(page)
                    print(i)
                    print("")
                    pass
                    
                try:
                    num_tel = etablissement.xpath('//div[contains(@class,"establishment--search_item__content")]/p/text()').extract()[i*3 + 1]
                except Exception as e: 
                    print("no telephone")
                    print(e)
                    print(page)
                    print(i)
                    print("")
                    pass
                
                try:
                    email = etablissement.xpath('//div[contains(@class,"establishment--search_item__content")]/p/a/@href').extract()[i]
                except Exception as e: 
                    print("no email")
                    print(e)
                    print(page)
                    print(i)
                    print("")
                    pass
                
                try: 
                    tags = Selector(text= etablissement.xpath('//div[contains(@class,"etablissement--search__item__tags")]').extract()[i]).xpath('//span/text()').extract()
                except Exception as e: 
                    print("no tags")
                    print(e)
                    print(page)
                    print(i)
                    print("")
                    pass
                    
                etablissement_details = {"nom": post_processing_string(nom_etablissement),
                                        "address": post_processing_string(address),
                                         "telephone": post_processing_string(num_tel,field = 'telephone'),
                                         "email": post_processing_string(email, field = 'email'),
                                         "tags": post_processing_string(tags, field = 'tags')}

                results[index] = etablissement_details
                index += 1 
                    
            # Moving on to the next page 
            page += 1
            
            # keep track
            if page%10 == 0 : 
                print("page #", page)
            
            # Stop at page 3200, as there aren't any 
            if page == 3200:
                return results 
                break
            
        except Exception:
            print(page)
            print(i)
            traceback.print_exc()
            return results            

resultats = scrap_etablissements_contact()
```

## resultats

```python
df_resultats = pd.DataFrame(columns = ['nom','address','telephone','email','tags'])

for i in resultats.keys():
    df_resultats = df_resultats.append(resultats[i],ignore_index=True)
```

```python
with open('resultats.txt', 'w') as outfile:
    json.dump(resultats, outfile)
```
